```python
# Plot distribution of power
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], bins=40, kde=True, color='blue')
plt.title('Distribution of Power Generated (kW)')
plt.xlabel('Power Generated (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bi-variate analysis
# Scatter plot with target feature

plt.figure(figsize=(20, 45))#, facecolor='yellow')
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'], colorizer='red')
    plt.title(f'{column} vs Power Generated kW')
    plt.xlabel(column)
    plt.ylabel('Power Generated (kW)')
plt.tight_layout()
plt.show()
```
